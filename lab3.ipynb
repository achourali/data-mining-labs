### Dividing data set to training set and testing set

```python

import random as random


def split1(data, target, training_percent):
    sample = list(zip(data, target))
    random.shuffle(sample)
    data, target = zip(*sample)
    learningLength = int(training_percent*len(data))
    dataS1 = data[:learningLength]
    targetS1 = target[:learningLength]
    dataS2 = data[learningLength:]
    targetS2 = target[learningLength:]
    return dataS1, targetS1, dataS2, targetS2
    
    
split1(irisData.data,irisData.target,0.6)
```

```python

def average_error_split1():
    sum = 0
    cycles = 1000
    for i in range(cycles):
        print(i)
        dataS1, targetS1, dataS2, targetS2 = split1(
            irisData.data, irisData.target, 2/3)
        nb_clf = naive_bayes.MultinomialNB(fit_prior=True)
        nb_clf.fit(dataS1, targetS1)

        predictions = nb_clf.predict(dataS2)

        sum += error(predictions, targetS2, nb_clf, dataS2)

        print(sum/i)


average_error_split1()
```

```python


from sklearn.model_selection import train_test_split

def average_error_train_test_split():
    sum = 0
    cycles = 1000
    for i in range(cycles):
        train_data, test_data, train_target, test_target = train_test_split(irisData.data, irisData.target, test_size=0.2)

        
        nb_clf = naive_bayes.MultinomialNB(fit_prior=True)
        nb_clf.fit(train_data, train_target)

        predictions = nb_clf.predict(test_data)
        
        

        sum += error(predictions, test_target, nb_clf, test_data)
        print(sum/i)
    
    
average_error_train_test_split()
```

#### Error estimation through cross-validation

```python
from sklearn.model_selection import cross_val_score
import numpy

nb_clf = naive_bayes.MultinomialNB(fit_prior=True)
for i in range(3,9):
  error = numpy.average(1-  cross_val_score(nb_clf,irisData.data,irisData.target,cv=i))
  print(i,error)
```
